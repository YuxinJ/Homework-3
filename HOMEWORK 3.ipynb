CREATE TEMP TABLE tempnaics AS(SELECT *
FROM naics WHERE naics LIKE '%----%')

SELECT *
FROM tempnaics
LIMIT 10; 

CREATE TEMP TABLE sector1 AS(SELECT substring(naics for 2)AS naics, naics_description
FROM tempnaics);

CREATE TEMP TABLE naics_temp2 AS(SELECT epa_handler_id, substring(naics_code for 2) AS naics_code
FROM hnaics); 

SELECT *
FROM naics_temp2
LIMIT 10;

SELECT naics_description, COUNT(*)
FROM naics_temp2
JOIN sector1 ON naics_temp2.naics_code = sector1.naics
GROUP BY naics_description
ORDER BY COUNT(naics_description);
--Answer:
--most handlers: Other Services(except Public Administration)
--least handlers: Management of Companies and Enterprises

FROM hsectors 
JOIN cmecomp3 ON hsectors.epa_handler_id = cmecomp3.handler_id
GROUP BY naics_description)

CREATE TEMP TABLE proportion1 AS(
SELECT sector_flag1.naics_description, count, tot_flag
FROM sector_flag1
JOIN tot_flag ON sector_flag1.naics_description = tot_flag.naics_description)
